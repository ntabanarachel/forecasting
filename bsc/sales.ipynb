```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading all sheets
sales = pd.read_excel('./datasets/Sales_project.xlsx')
customers = pd.read_excel('./datasets/Sales_project.xlsx', 'Customers Sheet')
sites = pd.read_excel('./datasets/Sales_project.xlsx', 'Site Location sheet')
products = pd.read_excel('./datasets/Sales_project.xlsx', 'Products_Sheet')
teams = pd.read_excel('./datasets/Sales_project.xlsx', 'Sales Team Sheet')
```

```python
sales.head(30)
```

```python
fig = (18,8)
plt.figure(figsize=fig)
heatmap = sns.heatmap(sales.corr(method='pearson'), annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='Black')
heatmap.set_title('correlation heatmap between variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
products_index = [a for a in range(len(products))]
products['products_index'] = products_index
products.set_index('products_index', inplace=True)
products.head()
```

```python
data = sales.join(customers.set_index('_CustomerID'), on='_CustomerID').join(sites.set_index('_SiteID'), on='_SiteID').join(products, on='_ProductID', how='left', lsuffix='_left', rsuffix='_right')
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()
data['SIte Name'] = enc.fit_transform(data['SIte Name'])
data['District'] = enc.fit_transform(data['District'])
data['Product Name'] = enc.fit_transform(data['Product Name'])
data['Item/Service Description'] = enc.fit_transform(data['Item/Service Description'])
data['Customer Names'] = enc.fit_transform(data['Customer Names'])
data['OrderDate'] = enc.fit_transform(data['OrderDate'])
```

```python
dataset = pd.DataFrame(data, columns=data.columns)
dataset
```

```python
fig = (18,8)
plt.figure(figsize=fig)
heatmap = sns.heatmap(dataset.corr(method='pearson'), annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='Black')
heatmap.set_title('correlation heatmap between variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```
